```python
for photo in photos:
    ids = pd.read_csv(f"D:/react-data/iconic/tsv/{photo}/txt/parsed-files-corrected.csv")
    ids = dict(zip(list([x.split('.')[0] for x in ids['path']]),list(ids['url'])))
    
    languages = pd.read_csv(f'C:/Users/ruben.ros/Documents/GitHub/iconicity/data/dataset/{photo}-sift-corrected.tsv',sep='\t')
    languages['language'] = [x.split('=')[0] for x in languages['language']]
    languages = dict(zip(list(languages['page_url']),list(languages['language'])))

    with open(f"D:/react-data/iconic/tsv/{photo}/metadata/dates.txt",'r') as f:
        dates = f.readlines()
        dates = [x.split('||') for x in dates]
        dates = {os.path.split(x[0])[-1].split('.')[0]:x[1][:4] for x in dates}

    idsn = {k:v for k,v in ids.items() if k in list(dates.keys())}
    if len(list(idsn.items())) != 0:
        print(photo,len(list(idsn.items())))
        texts = {}

        for id_,url in ids.items():
            try:
                file_path = os.path.join(f"D:/react-data/iconic/tsv/{photo}/txt/files_corrected",id_ + '.txt')
                if os.path.exists(file_path) == False:
                    file_path = os.path.join(f"D:/react-data/iconic/tsv/{photo}/txt/files",id_ + '.txt')
                if os.path.exists(file_path) == False:
                    print(file_path,"path does not exist")
                text = Data.rt(file_path)
                date = dates[id_]
                lan = languages[url]

                if lan == "en":
                    id_ = '_'.join([id_,date])
                    texts.update({id_:text})
            except Exception as e:
                continue

        print('Left with n. of English texts:',len(texts.keys()))

        texts = {k:Data.Tokenize(v) for k,v in tqdm(texts.items())}


        # Import Model Texts and Match
        json_ = glob.glob('D:/react-data/iconic/models/*.json')
        json_ = [x for x in json_ if photo in x][0]

        with open(json_,'r') as f:
            model_texts = json.load(f)

        matches = {}
        for id_,txt in texts.items():

            for mt in model_texts:
                if mt[0] == txt:
                    matches.update({mt[1][0]:id_})

        dfm = pd.DataFrame(list(matches.items()),columns=['doc2vec_index','identifier'])
        dfm['date'] = [x.split('_')[1] for x in dfm['identifier']]
        dfm['url'] = [idsn[x.split('_')[0]] for x in dfm['identifier']]
        dfm['id'] = [x.split('_')[0] for x in dfm['identifier']]
        dfm = dfm[["doc2vec_index","id","date","url"]]
        dfm.to_csv(f'C:/Users/ruben.ros/Documents/REACT/data/doc2vec/metadata/{photo}-metadata.csv',index=False)
```
