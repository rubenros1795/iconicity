```python
import os,glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
start_years = pd.read_csv('C:/Users/ruben.ros/Documents/ICON-ZENODO/results/diachronic-plots/start_years.csv')
start_years = dict(zip(list(start_years['photo']),list(start_years['year'])))
```

```python
for csv in glob.glob('C:/Users/ruben.ros/Documents/ICON-ZENODO/results/diachronic-plots/*.csv'):
    if "start" in csv:
        continue
    d = pd.read_csv(csv)
    continue
    start_year = start_years[os.path.split(csv)[-1].split('-')[0]]
    n = len(d.columns) - 1
    d = d.melt(id_vars='year')
    d.columns = ['year','cluster','value']
    d = d[d['year'] > start_year]
```

```python
list(d.columns)
```

```python
sns.set_style("darkgrid")

#sns.color_palette("tab20", 10)

f = sns.factorplot(data= d, 
               x= 'year',
               y= 'value',
               hue= 'cluster',
               lw=8, 
               ci=None, 
               capsize=.8, 
               palette=sns.color_palette("tab20", n),
               legend_out=False)
# f.fig.suptitle('Diachronic Normalized Cluster Prominence', fontsize= 25)
f.fig.set_size_inches(20, 10)
f.set_ylabels('Normalized Topic Prominence', fontsize= 20)
f.set_xlabels('Years', fontsize= 20)
f.set_xticklabels(fontsize= 15)
plt.legend(bbox_to_anchor=(1, -0.1), ncol=1,fontsize=16)
plt.savefig(f'date_matches/{photo}-diachronic-prominence.png',dpi=250, bbox_inches='tight')
plt.show()
```
